#### Ironhack Airlines 

Often Airlines sell more tickets than they have seats available, this is called overbooking. Consider the following:
- A plane has 450 seats. 
- Based on historical data we conclude that each individual passenger has a 3% chance of missing it's flight. 

If the Ironhack Airlines routinely sells 460 tickets, what is the chance that they have a seats for all passenger?

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import poisson
from scipy.stats import geom
from scipy.stats import expon
from scipy.stats import norm
```

```python

n = 460
p = 0.97

binom_dist = binom (n,p)

binom_dist.cdf(450)
```

```python
#chance of overbooking:
1-binom_dist.cdf(450)
```

**Challenge 3**

#### Ironhack Website

Consider a scenario related to Ironhack website traffic. Where:
- our website takes on average 500 visits per hour.
- the website's server is designed to handle up to 550 vists per hour.


What is the probability of the website server being overwhelmed?

```python
#code here
mu = 500

poisson_dist = poisson(mu)

1 - poisson_dist.cdf(550)
```

What is the probability of being overwhelmed at some point during a day? (consider 24hours)

```python
1-(1-p)**24
```

```python

p = (1 - poisson_dist.cdf(550))

n = 24

binom_dist = binom(n,p)


1-binom_dist.cdf(0)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, binom

# Parameters
mu = 500
limit = 550

# Poisson distribution to calculate p (probability of being overwhelmed in one hour)
poisson_dist = poisson(mu)
p = 1 - poisson_dist.cdf(limit)

# Number of hours range (from 1 to 24)
hours = np.arange(1, 25)

# Calculate the probability of being overwhelmed at least once for each hour
prob_overwhelmed = [1 - binom(n, p).cdf(0) for n in hours]

# Plotting the likelihood of being overwhelmed vs hours
plt.plot(hours, prob_overwhelmed, marker='o', linestyle='-', color='r')
plt.title('Probability of Overwhelm vs Hours')
plt.xlabel('Number of Hours')
plt.ylabel('Probability of Overwhelm')
plt.grid(True)
plt.show()
```

**Challenge 4**

#### Ironhack Helpdesk

Consider a scenario related to the time between arrivals of customers at a service desk.

On average, a customers arrives every 10minutes.

What is the probability that the next customer will arrive within the next 5 minutes?

```python

#the parameter represents the rate of events occuring per unit of time
mean_time = 10
lambda_inv = 1/mean_time

expon_dist = expon(scale = 1/lambda_inv)

expon_dist.cdf(5)
```

If there is no customer for 15minutes, employees can that a 5minutes break.

What is the probability an employee taking a break?

```python
1 - expon_dist.cdf(15)
```

**Challenge 5**

The weights of a certain species of birds follow a normal distribution with a mean weight of 150 grams and a standard deviation of 10 grams. 

- If we randomly select a bird, what is the probability that its weight is between 140 and 160 grams?

```python
mean = 150
std = 10

norm_dist = norm(loc = mean, scale = std)
norm_dist.cdf(160) - norm_dist.cdf(140)
```

```python
norm_dist.pdf(150)
#calculates the PROBABILITY DENSITY, meaning how likely it is to fall close to 150, since in a continuous distribution
#the probability of hitting EXACTLY 150 is technically 0, but the density is higher than around say 120, since it is 
#farther from the mean.
```

```python
norm_dist.pdf(120)
```

**Challenge 6**

If the lifetime (in hours) of a certain electronic component follows an exponential distribution with a mean lifetime of 50 hours, what is the probability that the component fails within the first 30 hours?

```python
mean_time = 50
lambda_inv = 1/mean_time

expon_dist = expon(scale = 1/lambda_inv)

expon_dist.cdf(30)
```
